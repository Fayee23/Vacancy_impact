## individual unit data

### Harrow

```python
import pandas as pd
import geopandas as gpd
#import geodatasets
```

```python
harrowUnit = pd.read_csv("data/Harrow/2023-05-05-E09000015-local.csv", index_col=False)
```

```python
harrowUnit.head(5)
```

```python
harrowUnit.tail(5)
```

```python
harrowUnit[harrowUnit['geometry'].isna()]
```

```python
harrowUnit['lat']=harrowUnit['geometry']
harrowUnit['lon']=harrowUnit['geometry']

for i in range(0, harrowUnit.shape[0]):
    harrowUnit.loc[i,'geometry']=harrowUnit.loc[i,'geometry'].replace('(', '')
    harrowUnit.loc[i,'geometry']=harrowUnit.loc[i,'geometry'].replace(')', '')
    harrowUnit.loc[i,'lat']=harrowUnit.loc[i,'geometry'].split()[-1]
    harrowUnit.loc[i,'lon']=harrowUnit.loc[i,'geometry'].split()[1]
    
harrowUnit=harrowUnit.drop(['geometry'], axis=1)
harrowUnit=gpd.GeoDataFrame(harrowUnit, geometry=gpd.points_from_xy(harrowUnit.lon, harrowUnit.lat))
harrowUnit
```

```python
harrowUnit.info()
```

```python
#save shapefile
harrowUnit.to_file('data/unitpoint/harrowUnit.shp')
```
